```python
#import library
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
#Baca data
boston = load_boston()
bos_ = pd.DataFrame(boston.data)
bos_.columns=boston.feature_names
bos_.head()
```

```python
bos_['PRICE'] = boston.target
print(bos_.head())
```

```python
#menentukan data dan target
x = bos_.drop('PRICE', axis = 1)
y = bos_['PRICE']
#split data ke data test dan data train
xtrain, xtest, ytrain, ytest=train_test_split(x, y, test_size=0.2)
```

```python
#SVR regression dengan kernel linear, rbf, dan polynomial
from sklearn.preprocessing import scale
from sklearn.svm import SVR
import time
x=scale(x)
y=scale(y)
xtrain, xtest, ytrain, ytest=train_test_split(x, y, test_size=0.15)

#SVR RBF
svr_rbftime=time.time()
svr_rbf = SVR(kernel='rbf', C=100, gamma=0.1, epsilon=.1)
svrrbf_fit = svr_rbf.fit (xtrain, ytrain)
ypred_2 = svr_rbf.predict(xtest)

#waktu yang dibutuhkan

print("waktu untuk SVR RBF: %.3f" % (time.time() - svr_rbftime)+'s')

x_ax = range(len(ytest))
plt.figure()
plt.scatter(x_ax, ytest,color='g', label="original")
plt.plot(x_ax, ypred_2, label="predicted")
plt.title("Boston test and predicted data")
plt.legend()
plt.show()

#SVR LINEAR
svr_lintime=time.time()
svr_lin = SVR(kernel='linear', C=100, gamma='auto')
svr_lin.fit (xtrain, ytrain)
ypred_3 = svr_lin.predict(xtest)
print("waktu untuk SVR Linear: %.3f" % (time.time() - svr_lintime)+'s')

x_ax1 = range(len(ytest))
plt.figure()
plt.scatter(x_ax1, ytest, color='red' , label="original")
plt.plot(x_ax1, ypred_3, label="predicted")
plt.title("Boston test and predicted data")
plt.legend()
plt.show()

#SVR POLY
svr_polytime=time.time()
svr_poly = SVR(kernel='poly', C=100, gamma='auto', degree=3, epsilon=.1,
               coef0=1)
svr_poly.fit (xtrain, ytrain)
ypred_4 = svr_poly.predict(xtest)
print("waktu untuk SVR Linear: %.3f" % (time.time() - svr_polytime)+'s')

x_ax2 = range(len(ytest))
plt.figure()
plt.scatter(x_ax2, ytest, color='red' , label="original")
plt.plot(x_ax2, ypred_4, label="predicted")
plt.title("Boston test and predicted data")
plt.legend()
plt.show()

```

```python
#Evaluation Model
from sklearn import metrics
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_validate, cross_val_score
from sklearn.metrics import mean_squared_error


score = svr_rbf.score(xtrain, ytrain)
cv_score = cross_val_score(svr_rbf, x, y, cv=10)
print("CV mean score: ", cv_score.mean())
mse = mean_squared_error(ytest, ypred_2)
print("MSE RBF: ", mse)
print("RMSE RBF: ", mse*(1/2.0))
print("R-squared RBF:", score)

print('-------------------------------')
score = svr_lin.score(xtrain, ytrain)
cv_score = cross_val_score(svr_lin, x, y, cv=10)
print("CV mean score lin: ", cv_score.mean())
mse = mean_squared_error(ytest, ypred_3)
print("MSE RBF lin: ", mse)
print("RMSE RBF lin: ", mse*(1/2.0))
print("R-squared lin:", score)
print('-------------------------------')
score = svr_poly.score(xtrain, ytrain)
cv_score = cross_val_score(svr_poly, x, y, cv=10)
print("CV mean score: ", cv_score.mean())
mse = mean_squared_error(ytest, ypred_4)
print("MSE RBF poly: ", mse)
print("RMSE poly: ", mse*(1/2.0))
print("R-squared poly:", score)
```
